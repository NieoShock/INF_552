### ii. Make boxplots for each of the independent variables. Use color to show Classes 0 and 1 (see ISLR p. 129).

```python
# _global.py 
CLASS0 = "class 0"
CLASS1 = "class 1"
EPSILON = 1e-8
```

```python
import numpy as np

def load_data(file_path):
	"""read records from csv file"""
	row_data = np.genfromtxt(file_path, dtype=None, delimiter=',', encoding='utf-8')[1:]
	X = np.array(row_data[:, :-1], dtype=float)
	Y_row_str_data = row_data[:, row_data.shape[1] - 1]
	y = np.array([1 if x == "Abnormal" else 0 for x in Y_row_str_data])
	# print(X)
	# print(y)
	return X, y
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
X_train, y_train = load_data('../assets/data.csv')
```

```python
labels = ['pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle', 'sacral_slope', 'pelvic_radius',
          'degree_spondylolisthesis']
results = ['Normal', 'Abnormal']

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 6), sharex=True)

normal_data = []
abnormal_data = []

for index in range(0, 6):
    normal_data.append(X_train[y_train == 0, index])
    abnormal_data.append(X_train[y_train == 1, index])

# box_color = colors[0]
square = dict(markerfacecolor='#e58e26', marker='s')
ax[0, 0].boxplot(normal_data, labels=labels, flierprops=square, whis=0.85, notch=True,
                 medianprops={'color': '#e58e26'},
                 whiskerprops={'color': "black"}, capprops={'color': '#e58e26'}, vert=False)
ax[0, 0].grid(True)
ax[0, 0].title.set_text(results[0] + " Symptom")
ax[1, 0].boxplot(abnormal_data, labels=labels, flierprops=square, whis=0.85, notch=True,
                 medianprops={'color': '#e58e26'},
                 whiskerprops={'color': "black"}, capprops={'color': '#e58e26'}, vert=False)
ax[1, 0].grid(True)
ax[1, 0].title.set_text(results[1] + " Symptom")
ax[0, 1].remove()
ax[1, 1].remove()
plt.xlim(0, 420)
plt.show()
```
